## Task 2:Dimensionality Reduction with PCA:

#### 1.Standardization of Features

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
print("Mean (approx):", np.round(X_scaled.mean(), 4))
print("Std (approx):", np.round(X_scaled.std(), 4))
```

#### PCA Implementation 

```python
from sklearn.decomposition import PCA
import numpy as np
```

```python
pca = PCA()
pca.fit(X_scaled)
# Scree plot
plt.figure(figsize=(11,7))
plt.plot(range(1, len(features)+1), pca.explained_variance_ratio_, marker='o')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()
```

### Cumulative explained variance

```python
plt.figure(figsize=(11,7))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Variance')
plt.show()
```

#### Transform Dataset Using Principal Components

```python
# Choose number of components (e.g., 3 for >80% variance)
n_components = 3
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)

print(f'PCA shape: {X_pca.shape}')
```

## Task 3:Clustering with Original Data

#### K-means Clustering on Original Data

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
labels_original = kmeans.fit_predict(X_scaled)
```

#### Clustering Results

```python
plt.figure(figsize=(11,8))
sns.scatterplot(x=X_scaled[:,0], y=X_scaled[:,1], hue=labels_original, palette='Set1')
plt.title('K-means Clusters (Original Data)')
plt.xlabel(features[0])
plt.ylabel(features[1])
plt.show()
```

#### Evaluates Clustering Performance

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

print('Silhouette Score:', silhouette_score(X_scaled, labels_original))
print('Davies-Bouldin Index:', davies_bouldin_score(X_scaled, labels_original))
```

## Task 4:Clustering with PCA Data

#### Clustering on PCA Dataset

```python
kmeans_pca = KMeans(n_clusters=3, random_state=42)
labels_pca = kmeans_pca.fit_predict(X_pca)
```

#### Visualize Clustering Results from PCA transformed data

```python
plt.figure(figsize=(11,8))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=labels_pca, palette='Set2')
plt.title('K-means Clusters (PCA Data)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

#### Clustering Performance (PCA)

```python
print('Silhouette Score (PCA):', silhouette_score(X_pca, labels_pca))
print('Davies-Bouldin Index (PCA):', davies_bouldin_score(X_pca, labels_pca))
```

## Task 5:Comparison and Analysis

#### Comparing Results and Metrics

```python
# Display clustering scores for both methods
print('Original Data Silhouette:', silhouette_score(X_scaled, labels_original))
print('PCA Data Silhouette:', silhouette_score(X_pca, labels_pca))
print('Original Data DaviesBouldin:', davies_bouldin_score(X_scaled, labels_original))
print('PCA Data DaviesBouldin:', davies_bouldin_score(X_pca, labels_pca))
```

## Task 6:Conclusion and Insights

### Summary
- PCA makes it easier to handle high-dimensional data by keeping most of the important information while using fewer variables.
- Clustering becomes simpler and often more effective when applied to data transformed by PCA, particularly if the original features are correlated.
- Applying PCA before clustering is recommended for large datasets with many features, as it reduces redundancy and helps focus on the main patterns in the data, though it may reduce the interpretability of individual features.
