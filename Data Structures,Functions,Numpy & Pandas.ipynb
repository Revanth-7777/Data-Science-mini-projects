### 1.Create a list of 5 integers.Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
# Create a list of 5 integers
numbers = [10, 20, 30, 40, 50]
print("Original list:", numbers)

# append()
numbers.append(60)
print("After append:", numbers)

# extend()
numbers.extend([70, 80])
print("After extend:", numbers)

# insert()
numbers.insert(2, 25)
print("After insert:", numbers)

# remove()
numbers.remove(40)
print("After remove:", numbers)

# pop()
popped = numbers.pop()
print("After pop:", numbers, "| Popped element:", popped)

# index()
index_of_25 = numbers.index(25)
print("Index of 25:", index_of_25)

# count()
count_20 = numbers.count(20)
print("Count of 20:", count_20)

# sort()
numbers.sort()
print("After sort:", numbers)

# reverse()
numbers.reverse()
print("After reverse:", numbers)
```

### 11.Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
# Create a small DataFrame manually
data = {
    'Name': ['Daya', 'Simhadri', 'Bheem', 'Tony', 'Abhi', 
             'Raghu', 'Devara', 'Sparrow', 'Jack', 'Kaboo'],
    'Age': [25, 30, 35, 40, 22, 28, 32, 27, 45, 30],
    'City': ['Hyd', 'Tadipatri', 'Thirupathi', 'Vizag', 'Tokyo', 
             'Mumbai', 'Chennai', 'Berlin', 'Brazil', 'Delhi'],
    'Salary': [50000, 55000, np.nan, 60000, 58000, 
               55000, 62000, np.nan, 70000, 55000]
}

df = pd.DataFrame(data)

print("=== Original DataFrame ===")
print(df)

# .info()
print("\n=== DataFrame Info ===")
print(df.info())

# .describe()
print("\n=== Statistical Summary ===")
print(df.describe())

# Select 'Name' and 'City' columns
print("\n=== Selected Columns: Name & City ===")
print(df[['Name', 'City']])

# Drop 'City' column
df_dropped = df.drop(columns=['City'])
print("\n=== DataFrame after Dropping 'City' Column ===")
print(df_dropped)

# Fill missing values in 'Salary' with mean
mean_salary = df['Salary'].mean()
df['Salary'] = df['Salary'].fillna(mean_salary)
print("\n=== DataFrame after Filling Missing 'Salary' Values ===")
print(df)

# Add a duplicate row intentionally for demonstration
df = pd.concat([df, df.iloc[[2]]], ignore_index=True)
print("\n=== DataFrame with Duplicate Row Added ===")
print(df)

# Remove duplicate rows
df_cleaned = df.drop_duplicates()
print("\n=== DataFrame after Removing Duplicates ===")
print(df_cleaned)
```
