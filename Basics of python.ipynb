## Exercise 1: Prime Numbers
Problem:  
Write a Python program that checks whether a given number is prime or not.  
A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

```python
def is_prime(num):
    # Prime numbers are greater than 1
    if num <= 1:
        return False
    # Check for factors
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True
number = int(input("Enter a number to check if it's prime: "))
if is_prime(number):
    print(f"{number} is a Prime Number.")
else:
    print(f"{number} is NOT a Prime Number.")
```

## Exercise 2: Product of Random Numbers
Problem:  
Generate two random numbers and ask the user to enter their product.  
Check if the user’s answer is correct and display an appropriate message.

```python
import random
num1 = random.randint(1, 20)
num2 = random.randint(1, 20)
print(f"Multiply the following numbers: {num1} × {num2}")
user_answer = int(input("Enter the product: "))
correct_answer = num1 * num2
if user_answer == correct_answer:
    print("Correct! Well done.")
else:
    print(f"Incorrect. The correct answer is {correct_answer}.")
```

## Exercise 3: Squares of Even or Odd Numbers 
Print the squares of all even OR odd numbers within the range 100 to 200.  
We’ll choose even numbers for this example.

```python
print("Squares of EVEN numbers from 100 to 200:\n")

for i in range(100, 201):
    if i % 2 == 0:
        print(f"{i}^2 = {i**2}")
```

## Exercise 4: Hospital Billing system
Problem Statement (Question): Hospital Billing System
A hospital charges its patients based on the following:
•	Room charges per day: ₹2,000
•	Doctor consultation fee (flat): ₹1,500
•	Lab test charges: Based on the number of tests taken. Each test costs ₹300
•	Medicine charges: Total cost of medicines provided
•	Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied
Write a Python program to calculate the final bill for a patient who:
•	Stayed for 4 days
•	Had 3 lab tests
•	Had medicine charges of ₹2,400


```python
#Problem:
#A hospital charges patients based on:
#Room charges per day: ₹2,000
#Doctor consultation fee (flat): ₹1,500
#Lab test charges: ₹300 per test
#Medicine charges: user-specified
#Discount: 10% if total before discount > ₹10,000

#Given details for this patient:
days_stayed = 4
num_tests = 3
medicine_charges = 2400

#Given Fixed charges
room_charge_per_day = 2000
doctor_fee = 1500
lab_test_charge = 300

#Step1:Total charges calculating before discount
room_total = days_stayed * room_charge_per_day
lab_total = num_tests * lab_test_charge

total_before_discount = room_total + doctor_fee + lab_total + medicine_charges

#Step2: Apply discount if neccessary
if total_before_discount > 10000:
    discount = 0.10 * total_before_discount
else:
    discount = 0

final_bill = total_before_discount - discount

#Step3:Final results
print("Hospital Billing Summary:")
print(f"Room Charges for 4 days (₹{room_charge_per_day}/day × {days_stayed} days): ₹{room_total}")
print(f"Doctor Consultation Fee: ₹{doctor_fee}")
print(f"3Lab Test Charges (₹{lab_test_charge} × {num_tests} tests): ₹{lab_total}")
print(f"Had Medicine Charges: ₹{medicine_charges}")
print("--------------------------------------")
print(f"Total Before Discount: ₹{total_before_discount}")
print(f"Discount Applied: ₹{discount}")
print("--------------------------------------")
print(f"Final Bill Amount: ₹{final_bill}")
```
